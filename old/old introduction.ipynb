# Week 2

## Introduction to Kinematics (15.8.2016)
We repeat the concepts of streamline and pathline and introduce on the streaklines. Examples will be given. The pathlines can be calculated by solving the ordinary differential equation, yet in many cases no analytical solutions are available and we have to use a computer to obtain a numerical one. A standard technique is called the [Euler method](notebooks/02a Euler%20Method.ipynb).

## Flow field decomposition (17.8.2016)
Using a lab based coordinate system is convenient but leads to the fact that the fluid particle's acceleration is not only a function of time, but of space, too. We derive the acceleration of a fluid particle at a point in lab space and name it [material derivative](notebooks/03 Material%20Derivative.ipynb). 
The [decomposition of a flow field](notebooks/04_Flow%20Field.ipynb) allows to quantify the deformation caused by linear straining, rotation, and shear.

# Week 9

The [Navier Stokes equation](notebooks/13_Navier Stokes Equation.ipynb) is derived by choosing a suitable relation between rate of strain and shear rate, the so-called consititutive equation for a Newtonian liquid. We don't yet solve the Navier Stokes Equation analytically, yet we can use a [numerical solver](notebooks/13a_Navier Stokes example.ipynb) to understand the effect of viscosity.
Next, we will derive the [Bernoulli Equations](notebooks/14_Bernoulli Equation.ipynb) from the N.S. equation. They allow us solving many flow problems mostly along a streamline. We solve the [collapse of a spherical void](notebooks/14a_Rayleigh_ODE.ipynb) as an example for an unsteady Bernoulli problem.
