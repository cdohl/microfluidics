# Stokes' Sphere

The flow around a slowly translating sphere with velocity $U_\infty$ and radius $R$ which is equivallent to the steady flow around a stationary sphere.

The Reynolds number is

$$Re=\frac{R\,U_\infty}{\nu}\tag{1}$$

Stokes (1851) was the first to solve the problem by only accounting for pressure and viscosity terms of the Navier Stokes Equation. We start with a stationary sphere with its center at the origin of the coordinate system.

The boundary conditions are

$$\left.\vb{u}\right|_{r=R}=\vb{0}\tag{2}$$

where $r=\sqrt{x_1^2+x_2^2+x_3^2}$. Besides the cartesian coordinate system we use also the spherical coordinate system $(r,\phi,\theta)$. For $r\rightarrow\infty$ we have

$$u_1=0,\quad u_2=0,\quad u_3=U_\infty \quad .$$


The Navier Stokes Equations in absence of inertial terms and in a spherical coordinate system are

$$
\pad{p}{r}=\mu \left[\ppad{u_r}{r}+\frac{1}{r^2}\ppad{u_r}{\theta}+
\frac{2}{r}\pad{u_r}{r}+\frac{\cot\theta}{r^2}\pad{u_r}{\theta}\\
-\frac{2}{r^2}\pad{u_\theta}{\theta}-\frac{2 u_r}{r^2}-
\frac{2\cot\theta}{r^2}u_\theta\right]\tag{3}
$$

$$
\frac{1}{r}\pad{p}{\theta}=\mu\left[\ppad{u_\theta}{r}+
\frac{1}{r^2}\ppad{u_\theta}{\theta}+\frac{2}{r}\pad{u_\theta}{r}+
\frac{\cot{\theta}}{r^2}\pad{u_\theta}{\theta}\\
+\frac{2}{r^2}\pad{u_r}{\theta}-\frac{u_\theta}{r^2\sin^2\theta}\right]\tag{4}
$$

and the continuity equation becomes

$$
\pad{u_r}{r}+\frac{1}{r}\pad{u_\theta}{\theta}+\frac{2 u_r}{r}+
\frac{u_\theta\cot\theta}{r}=0\quad.\tag{5}
$$

Please note that we did not write an equation for the $u_\phi$-velocity as we ignore swirl, i.e. $u_\phi=0$. 


The boundary conditions in the spherical coordinate system are

\begin{eqnarray}
r=R &:& u_r(R,\theta)=0 \quad \mathrm{and} \quad u_\theta(R,\theta)=0\tag{6a}\\
r\rightarrow\infty &:& u_r\rightarrow U_\infty\cos\theta\quad\mathrm{and}\quad
u_\theta\rightarrow - U_\infty\sin\theta\tag{6b}
\end{eqnarray}

We have to solve the PDEs (3-5) with these 4 boundary conditions in a spherical coordinate system. We assume that the solution is separable, e.g. $u_r(r,\theta)=f(r)\,g(\theta)$.
The boundary conditions suggest

\begin{eqnarray}
u_r(r,\theta)&=& B(r)\cos\theta\tag{7a}\\
u_\theta(r,\theta)&=& -A(r)\sin\theta\tag{7b}\\
p(r,\theta)&=&\mu\,C(r)\cos{\theta}\tag{7c} 
\end{eqnarray}

The boundary conditions lead to 

$$A(R)=0,\quad B(R)=0,\quad A(\infty)=U_\infty, 
\quad \mathrm{and}\quad B(\infty)=U_\infty\tag{7d}$$.

If we insert Ansatz Eqs. (7a,b) into the continuity Eq. (5) we obtain

$$\fud{B}{r}+\frac{2(B-A)}{r}=0\tag{8}$$

and insert Eqs. 7(a,b,c) into Stokes Eq. (3) we obtain

$$\fud{C}{r}=\ffud{B}{r}+\frac{2}{r}\fud{B}{r}-\frac{4(B-A)}{r^2}\tag{9}$$

and insert Eqs. 7(a,b,c) into Stokes Eq. (4) we obtain

$$C(r)=\ffud{A}{r}+\frac{2}{r}\fud{A}{r}+\frac{2(B-A)}{r^2}\tag{10}\quad.$$


We can solve Eq. (8) for $A$ 

$$A=\frac{1}{2}r\fud{B}{r}+B\tag{11}$$

and insert into Eq. (10) to obtain

$$C=\frac{1}{2}r^2\fffud{B}{r}+3r\ffud{B}{r}+2\fud{B}{r}\tag{12}\quad .$$

We can now take the derivative of Eq. (12) with respect to $r$ and insert the resulting expression into Eq. (9) to obtain

$$
r^3 \ffffud{B}{r} + 8 r^2 \fffud{B}{r} + 8r \ffud{B}{r} - \fud{B}{r} = 0\tag{13}\quad .
$$

Multiplying Eq. (13) with $r$ results to the [Euler differential equation](https://en.wikipedia.org/wiki/Cauchy%E2%80%93Euler_equation) which can be solved by inserting the trial solution $B=r^k$ (or using [Wolfram Alpha](https://www.wolframalpha.com/input/?i=x%5E4+d4f%2Fdx4%2B8*x%5E3+d3f%2Fdx3%2B8*x%5E2+d2f%2Fdx2-8x+df%2Fdx%3D0) ). We then obtain an equation for $k$

$$k(k-1)(k-2)(k-3)+8k(k-1)(k-2)+8k(k-1)-8k=0\tag{14a}\quad ,$$

Which can be simplified to 

$$k(k-2)(k+3)(k+1)=0\tag{14b}\quad .$$

Thus the general solution for $B(r)$ is

$$B(r)=\frac{C_1}{r^3}+\frac{C_2}{r}+C_3+C_4r^2\tag{15}\quad .$$

From Eq. (11) we obtain

$$A(r)=-\frac{C_1}{2r^3}+\frac{C_2}{2r}+C_3+2C_4r^2\tag{16}$$

and from Eq. (12)

$$C(r)=\frac{C_2}{r^2}+10 C_4 r\tag{17}\quad .$$

Using the boundary conditions Eq. (7d) we can solve for the 4 constants and obtain

$$
C_1=\frac{1}{2}U_\infty R^3,\quad C_2=-\frac{3}{2}U_\infty R,\quad C_3=U_\infty,\quad C_4=0\quad .\tag{18}
$$


Thus our velocity field is (see Eq. (7a-c)):

\begin{eqnarray}
u_r(r,\theta)&=&U_\infty \cos \theta\left[ 1-\frac{3}{2}\frac{R}{r}+\frac{1}{2}\frac{R^3}{r^3}\right]\tag{19a}\\
u_\theta(r,\theta)&=&-U_\infty \sin \theta\left[ 1-\frac{3}{4}\frac{R}{r}-\frac{1}{4}\frac{R^3}{r^3}\right]\tag{19b}\\
p(r,\theta)&=&-\frac{3}{2}\mu\frac{U_\infty R}{r^2}\cos \theta\tag{19c}
\end{eqnarray}

Because we have omitted the inertia terms on the L.H.S. of the N.S. equations the solution is only valid for sufficiently small $Re$-numbers, i.e. $Re\ll1$.

To calculate the viscous force, $\vb{F}_{vd}$, on the sphere we need to multiply the deviatoric stress tensor with the normal vector of the surface element and integrate over the surface, thus:

$$
\vb{F}_{vd}=\oint_S \tb{\sigma} \, \vb{e}_r \dd S=\oint_S \begin{pmatrix}\sigma_{rr}\\\sigma_{r\theta}\\\sigma_{r\phi}\end{pmatrix}\dd S \tag{20}
$$

We therefore only need to investigate the components $\sigma_{rr}$, $\sigma_{r\theta}$, $\sigma_{r\phi}$. These can be obtained for example from Appendix B of Kundu & Cohen textbook and are:

\begin{eqnarray}
\sigma_{rr}&=&2\mu \pad{u_r}{r}\tag{21a}\\
\sigma_{r\theta}&=&\mu\left[\frac{1}{r}\pad{u_r}{\theta}+\pad{u_\theta}{r}
-\frac{u_\theta}{r}\right]\tag{21b}\\
\sigma_{r\phi}&=&\mu\left[\frac{1}{r\sin \theta}\pad{u_r}{\phi}+
r\pad{}{r}\left(\frac{u_\phi}{r}\right) \right]\tag{21c}\quad .
\end{eqnarray}

The stresses have to be evaluated at the surface $r=R$. It can be easily shown that 

$$\left.\sigma_{rr}\right|_{r=R}=0$$

Also as $u_r\ne u_r(\phi)$ and $u_\phi=0$ the stress $\sigma_{\phi r}=0$ everywhere. The remaining component is $\sigma_{r\theta}(r=R)$. We have three terms in Eq. (21b). The last term is zero because $u_\theta(r=R)=0$. The same for the first term where the $\partial u_r/\partial \theta(r=R)$ vanishes. The second term is 


$$\left. \pad{u_\theta}{r} \right|_{r=R}=U_\infty \sin \theta \left[\frac{3}{4} \frac{R}{r^2}+\frac{3}{4}\frac{R^3}{r^4}\right]=
\frac{3}{2}\frac{U_\infty\sin\theta}{R}\tag{22}\quad .
$$

Equation (22) provides the only remaining viscous stress contribution on the sphere. Yet, also the pressure distribution on the surface may lead to a net force

$$
p(r=R,\theta)=-\frac{3}{2}\frac{\mu U_\infty}{R}\cos \theta\tag{23}
$$

The total force in the direction of the flow, the so-called drag force, is

$$
F_D=\oint_S \left(-p\cos \theta+\sigma_{r\theta}\sin \theta\right)\dd S\tag{24}\quad,
$$

where we used the $\sin$ and $\cos$ to project the normal and tangential component of the stress in the direction of the incoming flow. 

We obtain

$$
F_D=\int_0^\pi\left(\frac{3}{2}\frac{\mu U_\infty}{R}\cos^2\theta +\frac{3}{2}
\frac{\mu U_\infty}{R}\sin^2 \theta\right) (2\pi R^2\sin\theta)\dd\theta\\
F_D=\int_0^\pi 3 \pi \mu U_\infty R \sin\theta \dd \theta=6\pi \mu U_\infty R\tag{25}
$$

which is the well know drag force on a sphere at very low Reynolds number.


## Flow field

The velocity field Eq. (19a,b) can be written in cartesian coordinates

\begin{eqnarray}
u_1&=&-\frac{3}{4}\frac{U_\infty x_1 x_3}{r^2}\left(1-\frac{R^2}{r^2}\right)\tag{26a}\\
u_2&=&-\frac{3}{4}\frac{U_\infty x_2 x_3}{r^2}\left(1-\frac{R^2}{r^2}\right)\tag{26b}\\
u_3&=&U_\infty\left(1-\frac{3}{4}\frac{R}{r}-\frac{1}{4}\frac{R^3}{r^3}\right)-
\frac{3}{4}\frac{U_\infty x_1^2}{r^2}\left(1-\frac{R^2}{r^2}\right)\tag{26c}
\end{eqnarray}

Below you find the flow field in the $(x_2,x_3)$-plane around a sphere. Note that the velocity is dropping to zero as the flow approaches the sphere.

```python
%matplotlib inline 
from scipy.special import erf 
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma #for the masked array
import math
from ipywidgets import interact
import ipywidgets as widgets

xstart=-3
xend=3 #start of grid in x2 and x3-direction
steps=20 #number of gridpoints
h=(xend-xstart)/(steps-1.) #gridspacing
phi=np.linspace(0.,2.*np.pi,100)
x2=np.linspace(xstart,xend,steps)
x3=np.linspace(xstart,xend,steps)
xx3,xx2=np.mgrid[xstart:xend:steps*1j,xstart:xend:steps*1j] #generate two grids (note the return order yy,xx)

def plotstokessphere(R):
    mu=1e-1
    rho=1e3
    U=1.
    r=np.sqrt(xx3**2+xx2**2)
    
    #there may be errors generated by the 1/r terms, let's ignore them
    np.seterr(divide='ignore', invalid='ignore')
    
    u3=U*(1-.75*R/r-0.25*(R/r)**3.) #Eq. (25b)
    u2=-.75*U*xx2*xx3/(r**2.)*(1-(R/r)**2) #Eq. (25c) for x1=0
    
    #We used masked arrays not to plot inside the sphere
    M = np.zeros(u2.shape, dtype=bool) 
    M[np.where(r<R)]=True
    u2=ma.masked_array(u2, mask=M)
    u3=ma.masked_array(u3, mask=M)
    
    uu = np.sqrt(u2**2+u3**2) #speed
    lw = (uu/uu.max()*2)**2. #scaled linewidth (thinner is slower)
    
    plt.figure(figsize=(7,7))
    #vector plot  
    plt.quiver(xx2,xx3,u2,u3,angles='xy')
    #sphere
    plt.plot(R*np.cos(phi),R*np.sin(phi),'k',linewidth=5)
    #streamline
    plt.streamplot(xx2,xx3,u2,u3, linewidth=lw, color=uu,cmap=plt.cm.plasma)

    plt.xlabel(r'$x_2$ (m)',size=18);plt.ylabel(r'$x_3$ (m)',size=18)
    plt.tick_params(labelsize=15)
    plt.axis([xstart,xend,xstart,xend])

interact(plotstokessphere, R=widgets.FloatSlider(min=.1,max=2,step=.1,value=0.5,\
                                            description='radius (m)'));
```

## Stokes settling velocity

The constant velocity at which a sphere settles in a viscous flow under the action of gravity is called the Stokes settling velocity, $U_S$. It is determined from a force balance with the weight, buoyancy, and drag:

$$
0= \rho_s \frac{4}{3}\pi R^3 g - \rho \frac{4}{3}\pi R^3 g - 6 \pi \mu R U_S\tag{27}\quad,
$$

where $\rho_s$ is the density of the shere and $g$ the gravitational acceleration.
The L.H.S. is zero as we assume a steady translation without acceleration (no inertia). Solving for $U_S$ we obtain

$$
U_S=\frac{2\left(\rho_s-\rho\right)R^2 g}{9\mu}\tag{28}
$$
