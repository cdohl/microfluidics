## Week 3: Intro to Microfluidics & Derivation of the Conservation of Mass

Soft mathematical introduction to fluid mechanics. We start with the [material derivative](Material%20Derivative,%20Gaussian%20Divergence%20Theorem%20and%20Conservation%20of%20Mass.ipynb) and also introduce the *Gaussian Divergence Theorem* and the *Conservation of Mass*. It is important that you understand the [Eulerian description](https://en.wikipedia.org/wiki/Lagrangian_and_Eulerian_specification_of_the_flow_field) of flows. We will start with the [Cauchy's Equation of Motion](Cauchy's%20Equation%20of%20Motion%20&%20Navier%20Stokes%20Equation.ipynb) for a continuum from Newton's 2nd Law of Motion for a deformable object.


## Week 4: Navier Stokes equation derivation and first analytical solutions

We insert the surface stresses for a linear fluid and through simplification obtain the Navier Stokes Equation for a [incompressible and constant viscosity fluid](Cauchy's%20Equation%20of%20Motion%20&%20Navier%20Stokes%20Equation.ipynb). Then we look at some solution of a [numerical flow solver](13a_Navier%20Stokes%20example.ipynb).


## Week 5: 
We discuss analytical solutions to the Navier Stokes Equation in [planar and cylindrical symmetry](Analytical%20solutions%20to%20steady%20and%20laminar%20flows.ipynb) and [Dimensional Analysis](Dimensional%20Analysis%20of%20the%20Navier%20Stokes%20Equation.ipynb).


## Week 6:

Navier Stokes solution to an [unsteady problem](Analytical%20Solution%20to%20an%20unsteady%20flow.ipynb) (Stoke's first problem) and comparison with the CFD solver. For inviscid flows, we simplify the Navier Stokes Eq. to the [Bernoulli equation](Bernoulli%20Equation.ipynb). 


## Week 7:

Application of the [Unsteady Bernoulli Equation](Example%20of%20unsteady%20Bernoulli%20equation.ipynb) for the collapse of a spherical void. Introduction to [potential flow](Ideal%20Flow%20&%20Solutions%20to%20the%20Laplace%20Equation.ipynb).


## Week 8
Derivation of the force on a sphere [in potential flow](Potential%20Sphere.ipynb) and in a viscous flow, i.e. the [Stokes Sphere](Stokes%20Sphere.ipynb) problem. We will also look at a viscous microfluidic flow with a potential flow solution, the [Hele Shaw flow](Hele%20Shaw%20Flow.ipynb).


## Week 9
Microfluidic systems (Lab on a Chip) are composed typically out of inlets, outlets, and one or many connecting channels. In this lecture we will develop a strategy to describe these connected channel networks with a hydraulic [Resistance](Hydraulic%20resistance.ipynb). 

## Week 10
We will look at the basic principle of diffusion and discuss several solutions of the diffusion equation in this [section](Diffusion.ipynb).


## Week 11
[Capillarity](Capillarity.ipynb) is the force acting at interfaces between solids, gases and liquids. Here we'll start from the basic concept of the Gibbs free energy to the Young-Laplace equation and the Young-Equation. We'll see that a surface tries to maintain minimum area within the constraints set. We'll take a look at the rise and rise velocity of a vertical microcapillary and the pumping of liquid of an initialy dry microfluidic channel. If you want to dive further I strongly recommend the book [Capillarity and Wetting Phenomena](https://link.springer.com/book/10.1007%2F978-0-387-21656-0) by  Pierre-Gilles de Gennes, Francoise Brochard-Wyart, Davide Quere.
