## Week 2: Intro to Microfluidics & Derivation of the Conservation of Mass

Soft mathematical introduction to fluid mechanics. We start with the [material derivative](Material%20Derivative,%20Gaussian%20Divergence%20Theorem%20and%20Conservation%20of%20Mass.ipynb) and also introduce the *Gaussian Divergence Theorem* and the *Conservation of Mass*. It is important that you understand the [Eulerian description](https://en.wikipedia.org/wiki/Lagrangian_and_Eulerian_specification_of_the_flow_field) of flows. We will start with the [Cauchy's Equation of Motion](Cauchy's%20Equation%20of%20Motion%20&%20Navier%20Stokes%20Equation.ipynb) for a continuum from Newton's 2nd Law of Motion for a deformable object.


## Week 3: Navier Stokes equation derivation and first analytical solutions

We insert the surface stresses for a linear fluid and through simplification obtain the Navier Stokes Equation for a [incompressible and constant viscosity fluid](Cauchy's%20Equation%20of%20Motion%20&%20Navier%20Stokes%20Equation.ipynb). Then we look at some solution of a [numerical flow solver](13a_Navier%20Stokes%20example.ipynb).

## Week 10
[Capillarity](Capillarity.ipynb) is the force acting at interfaces between solids, gases and liquids. Here we'll start from the basic concept of the Gibbs free energy to the Young-Laplace equation and the Young-Equation. We'll see that a surface tries to maintain minimum area within the constraints set. We'll take a look at the rise and rise velocity of a vertical microcapillary and the pumping of liquid of an initialy dry microfluidic channel.
