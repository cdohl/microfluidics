## Capillarity

We need Gibbs free energy $\cal G$, the energy of a system with variables pressure $p$, temperature $T$, and particle number $N$. The total Gibbs energy $\cal G$ of a system consisting of two sub-subsystems is then the sum of the two energies of the sub-system $\cal G_i$ plus the energy of the surface. Now we describe the surface as a function of a variable $\xi$, e.g. position, volume, or shape. At equilibrium (lowest energy), the surface takes the shape $\xi=\xi_0$. Variations of $\xi=\xi_0+\delta\xi$ must result in a vanishing variation $\delta \cal G$ of the free energy, otherwise the system would jump to a new lower free energy with a new $\xi_0$.

$$\delta G= \pad{\cal G}{\xi}\delta\xi=\left(\sum_i\pad{\cal G_i}{\xi}\right)\delta\xi=0\tag{1}$$

This equation will be used to establish governing equations for capillary effects.

### Surface tension

<img src="files/pics/surface_tension.png" width=600> 

The origin of surface tension for a liquid-gas interface lies in the chemical bonds between neighboring molecules. The molecule in the neighborhood of a surface misses the chemical bonds in the dirction of the surface. Bringing a molecule twoards the surface requires to "cut" off these bonds which costs energy. The molecules at an interface have a higher energy which is termed surface energy. Let us estimate this energy per area:

$$[\gamma]=\text{J}\,\text{m}^{-2}=\text{N}\,\text{m}^{-1}$$ 

A water molecule within the liquid has 6 nearest neighbours (faces of a cube), while at a surface only 5 neightbours. The area of a single water molecule is $A=(0.3\text{nm})^{2}$ and the inter-molecular hydrogen bond $\Delta E\approx 2 k_B\,T\approx 50\,\text{meV}$. Thus

$$\gamma\approx\frac{2 k_B T}{A}\approx\frac{50\,\text{meV}}{(0.3\,\text{nm})^2}\approx\frac{50\cdot10^{-3}\,1.6\cdot10^{-19}\,\text{C}\,\text{V}}{(0.3\,\text{nm})^2}\approx90\,\text{mJ}\,\text{m}^{-2}$$

This values is not too far off from the measured value of $\gamma=70\,\text{mJ}/\text{m}^{-2}$.

We can define the surface energy from the Gibbs free energy as

$$\gamma\equiv\left(\pad{\cal G}{A}\right)_{p,T} \quad .$$

The surface energy can also be interpreted as a force per length having the unit $\text{N}/\text{m}=\text{J}/\text{m}^2$.

<img src="pics/surface_work.png" width=400px>

Consider a flat rectangular surface of width $l$ and length $X$. If we keep the width constant and stretch the surface the amount $\text{d}x$, then an external force must act to supply the work $\cal G=F\,\text{d}x$ necessary for creating the surface $l\,\text{d}x$ containing the energy $\Delta {\cal G}=\gamma\,l\,\text{d}x$, or

$$\frac{\Delta \cal G}{A}=\frac{\gamma\,l\,\text{d}x}{\text{d}x\,l}=\gamma\quad .$$


### Curved surface and Young-Laplace pressure

<img src="pics/laplace_pressure.png" width=400px>


Consider a small piece of a curved surface with area $A=\delta x\,\delta y$ in equilibrium separating two regions using Gibbs free energy formulation. Let us now expand the area through a small displacement $\delta z$ parallel to the normal vector of the surface. The two radii of curvature $R_i$ with $i=1,2$ of the surface thus expand from $R_i$ to $R_i+\delta z=(1+\delta z/R_i)\,R_i$. The two arc lengths $\delta x$ und $\delta x$ enlarge to $\delta x\left(1+\frac{\delta z}{R_1}\right)$ and $\delta y\left(1+\frac{\delta z}{R_2}\right)$. Now neglecting terms of $(\delta z)^2$ the area of the surface is enlarged by

$$\delta A\approx \left(\frac{\delta z}{R_1}+\frac{\delta z}{R_2}\right)\,A$$

Now consider the aread $A$ covers a volume $V$. In absence of gravity there are two contributions to the change $\delta G$ of the free energy of the system: increase in the surface energy ${\cal G}_\text{surf}$ due to an increased in area, and a decrease of pressure-volume energy ${\cal G}_\text{p V}$ due to an increase in volume:

$$\delta{\cal G}=\delta {\cal G}_\text{surf}+\delta {\cal G}_\text{p V}=\gamma\,\delta A-A\,\delta_z\,\Delta p_\text{surf}=0 \tag{2} .$$

The difference in pressure between the inner and the outer surface is called the Young-Laplace pressure

$$\Delta p_\text{surf}=\left(\frac{1}{R_1}+\frac{1}{R_2}\right)\gamma\tag{3}$$

### Contact angle

When a liquid contacts a solid and is partially surrounded by a vapor/gas phase it develops at the three phase contact a specific angle, named the contact angle $\theta$. It is the angle betwen the solid-liquid and the liquid-gas interface, see below figure. The contact angle is well defined for a static droplet, yet once it starts moving contact lines and the associated contact angles are still an active area of research.

<img src="pics/contact_angle.png" width=400px>

Besides the liquid-gas interface ($\gamma_\text{lg}$), also the wetted solid interface ($\gamma_\text{sl}$) and that of solid-gas interface ($\gamma_\text{sg}$) posses interfacial energy. We can derive an expression for the static contact angle that accounts for these three interfacial energies.

Consider the three phase contact line in the figure below is moved (e.g. by infitesial tilting) to the right such that the contact lines establishes a new location where the $\delta \cal G=0$. The interface has then moved by a distance $\delta l$. The free energy per depth $w$ is then:

<img src="pics/derivation_young.png" width=400px>

$$\frac{1}{w}\, \delta {\cal G}=\gamma_\text{sl}\,\delta l+\gamma_\text{lg}\,\delta l\,\cos\theta-\gamma_\text{sg}\,\delta l=0\quad .\tag{4}$$ 

Which can be written as

$$\cos\theta = \frac{\gamma_\text{sg}-\gamma_\text{sl}}{\gamma_\text{lg}}\quad .\tag{5}$$

Solids where $\theta<90^\circ$ for water are called water loveing or hydrophilic, and for $\theta>90^\circ$ are water fearing or hydrophobic.

### Capillary rise

In thin capillarities liquid can rise through the minimization of the Gibbs free energy, reducing surface energy into potential energy of the liquid column

$${\cal G}=\text{min} \left( \gamma \int_{\partial \Omega}\dd a+ \rho\, g\int_\Omega z \dd \vb{r}\right)\tag{6}$$

under the constrain that the volume remains constant. From Eq. (6) we can derive a length scale at which surface energy and potential energy balance. This length scale is named capillary length $l_\text{cap}$

$$l_\text{cap}\equiv \sqrt{\frac{\gamma}{\rho \, g}}\tag{7}$$

which for water-air interface takes the value

$$l_\text{cap}^\text{water-air}\approx 2.7\,\text{mm}\quad .$$

That means, that for a length $a\ll l_\text{cap}$ gravity does not influence the shape of the free interface (free interfaces are interfaces between a liquid and a gas). Thus droplets sitting on substrate and having a diameter of say $1\,$mm possess a spherical cap shape, thus their shape is not affected by gravity.

<img src="pics/capillary_rise.png" width=400px>


Let us first derive an expression for a capillary rise in a microscopic vertically aligned tube with a radius $a\ll l_\text{cap}$. Therefore, the shape of the liquid-air interface will be part of a sphere with a radius 

$$R=\frac{a}{\cos\theta}\quad ,$$

see above figure. There will be a pressure drop across the interface, such that the pressue in the gas is $\Delta p_\text{surf}$ larger than in the liquid below. Thus the pressure in the liquid is

$$p_\text{liq}(H) = p_0 - \Delta p_\text{surf}=p_0 - \frac{2\gamma}{R}=p_0-\frac{2\gamma}{a}\,\cos\theta\tag{8}.$$

The pressure in the liquid far from the tube is $p_0$, at the same height $z=0$ but within the tube the pressure is still $p_0$. Now in the tube, the pressure from $p_\text{liq}(H)$ at the miniscus increases due to the hydrostatic pressure. 

$$p_0=p_\text{liq}(z=0)=p_\text{liq}(z=H)+\rho g H\tag{9}$$


Inserting the Laplace pressure in Eq. (9) from Eq. (9) and solving for $H$ allows calculating the capillary rise

$$H=\frac{2\gamma}{\rho g a }\cos \theta=2 \frac{l_\text{cap}^2}{a} \cos \theta = \frac{2}{\rho g a} \left(\gamma_\text{sg}-\gamma_\text{sl}\right)\tag{10}\quad .$$

From the height $H$ we can estimate the coefficient of surface tension $\gamma$

$$\gamma=\frac{\rho g}{2} \frac{a \, H}{\cos \theta}\tag{11}\quad .$$


### Capillary rise dynamics in vertical tubes

We can derive an expression for the dynamics of the rise after having established the capillarity in a thin tube and its balance with the hydrstatic pressure. The only additional ingredient is the well known Poiseuille flow in a tube. Assuming a fully developed flow


$$\fud{L(t)}{t}= U(t)=\frac{Q(t)}{A}=\frac{Q(t)}{\pi a^2}=\frac{a^2\,\Delta p(t)}{8\mu}\frac{1}{L(t)}\tag{12}\quad .$$

The pressure drop driving the Poiseuille flow is pressure difference between the bottom $p_0$ and the top $p_0-\Delta p_\text{surf}+\rho g L(t)$. Thus 

$$\Delta p(t)=p_0 - \left(p_0 - \Delta p_\text{surf} + \rho g L(t) \right)=\Delta p_\text{surf}-\rho g L(t)\quad .\tag{13}$$

We can now insert Eq. (13) into Eq. (12) while using the expression $p_\text{surf}$ from Eq. (8) we obtain

$$\fud{L(t)}{t}=  \frac{\gamma}{8\,\mu}\left( 2a\cos\theta\frac{1}{L(t)}-\frac{\rho g a^2}{\gamma}\right) = \frac{\rho g a^2}{8\,\mu}\left( \frac{H}{L(t)} -1 \right)\quad . \tag{14}$$

Eq. (14) is an ordinary differntial equation (ODE). Before solving it, we make the equation dimensionless. For the length $L$ we can use the height $L'=L/H$ and for the time we see that the factor in front of the brackets in Eq. (14) is a velocity, thus multiplied with a length scale $H$ we have a time scale, the capillary time scale:

$$\tau_\text{cap}=\frac{8\,\mu}{\rho g a^2 H}$$

and $t'=t/\tau_\text{cap}$.

Let us now rewrite the ODE (14) in non dimensional variables

$$\fud{L'(t')}{t'}=\frac{1}{L'(t')}-1\tag{15}\quad ,$$

with the initial condition is $L'(t'=0)=0$ and we expect a final state of $L'(t'\rightarrow\infty)=1$.

This is a nonlinear ODE and obtaining a general solution is difficult or may not exist. Therefore we simiplify this equation by looking at the limits a) of early times $t'\ll 1$ and b) $t'\rightarrow \infty$.

For the early times $t'\ll 1$:

$$\fud{L'(t')}{t'}=\frac{1}{L'(t')}$$

which can be simply integrated and we obtain 

$$L'(t')=\sqrt{2\,t'}\quad t'\ll 1\tag{16}$$

and for the late times $t'\rightarrow \infty$ we can rewrite $L'=1-\delta L'$, where $0\leq \delta L\ll 1$. Inserting this into Eq. (15) we obtain


$$-\fud{\delta L'(t')}{t'}=\frac{1}{1-\delta L'(t')}-1\approx\delta L'(t')$$

for which the solution is $\delta L'(t')=\exp (-t')$ and we obtain for $L'$

$$L'(t')=1-e^{-t'}\quad t'\gg 1\tag{17}\quad .$$

The solution and the early time scaling is plotted in the figure below. Please check the program.

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def cap_rise(L, t):
    dLdt = 1./L-1.
    return dLdt


L0 = 1e-5 #small but not zero, why?

t = np.linspace(0.,5., 20000)
sol = odeint(cap_rise, L0, t)

plt.plot(t, sol[:,0],'k',label='solution')
plt.plot(t, np.sqrt(2.*t),'k--',label=r"$\sqrt{2 t'}$")

plt.ylim((0,1.1))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel(r"$t'$",fontsize=18)
plt.ylabel(r"$L'$",fontsize=18)
plt.legend(fontsize=15);
```

### Capillary rise dynamics in horizontal tubes: Capillary pumps


<img src="pics/capillary_pump.png" width=600px>

In microchannels the tube is placed horizontally, thus the gravitational force does not balance the capillary forces. The flow will continue as long as the length of the channels. This way liquid can be pumped into dry channel. We can derive the speed of pumping simialr to the vertical rise. Now we assume the channel to have rectangular cross section with a height much smaller than the width $h\ll w$ and $Q=h^3\,w\,\Delta p/(12\mu L)$. The driving pressure $\Delta p$ is now constant and is simply 

$$\Delta p=\Delta p_\text{surf}=\frac{\gamma}{h}\cos \theta$$

The speed of the flow is again calculated by assuming a fully developed Poiseuille flow in the whole channel

$$\fud{L(t)}{t}=U(t)=\frac{Q(t)}{A}=\frac{Q(t)}{w\,h}=\frac{h^2\,\Delta p_\text{surf}}{12\,\mu}\,\frac{1}{L(t)}\tag{18}$$

The equation can simply be integrated and gives

$$L(t)=\sqrt{2\,C\,t}\quad ,\tag{18a}$$

where $C=h^2\,\Delta p_\text{surf}/(12\,\mu)$ is the constant in Eq. (18). The characteristic time is $\tau_\text{adv}=h^2/(2C)$ or

$$\tau_\text{adv}\equiv \frac{6\,\mu}{\Delta p_\text{surf}}$$ 

and the solution to Eq. (18) is

$$L(t)=h \sqrt{ \frac{t}{\tau_\text{adv}}}\tag{19}$$

which is a similar result to the short time behaviour of the capillary rise problem. Why? Please compare Eq. (18a) with the diffusion equation.
