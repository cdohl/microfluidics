## Hydraulic Resistance

The [Poiseuille flow](Analytical%20solutions%20to%20steady%20and%20laminar%20flows.ipynb) (pressure driven incompressible steady flow through a straight channel) results in a constant flow rate $Q$ for a given pressure drope $\Delta p$, e.g. for a flow through a cylinder of length $L$. 

$$Q=\frac{\pi a^4}{8\mu}\frac{\Delta p}{L}\quad .$$
The equations for other channel shapes such as ellipic channels or rectangular channels have a similiar relationship. The flow rate is a linear function of the pressure drop. Thus we can write 

$$\Delta p=R_{\text{hyd}}\,Q\quad ,$$

where we have introduced the $R_{\text{hyd}}$ as the hydraulic resistance, similar to Ohms' law ($\Delta V= R \, I$). This relation is called the *Hagen Poiseuille law*. With the SI units for 

$$[Q]=\frac{\text{m}^3}{\text{s}}, \quad [\Delta p]=\text{Pa}$$

we obtain for the hydraulic resistance

$$[R]=\frac{\text{Pa}\,s}{\text{m}^3}\quad.$$



### Hydraulic resistance of channels connected in series


<img src="pics/channels_inseries.png" width=600px>

Consider placing two channels with hydraulic resistance $Q_1$ and $Q_2$ in series, such as shown in the figure below. Assuming the that Hagen-Poiseuille law is valid for each part (thus ignore entrance/exit flows) we can add the presure drops and obtain for the total resistance:

$$R=R_1+R_2\tag{14}$$

Entrance/exit effects can only be ignored at very low Reynolds numbers, $Re\ll 1$.


### Hydraulic resistance of channels connected in parallel


<img src="pics/channels_inparallel.png" width=600px>

Here we add the two flow rates, i.e. $Q=Q_1+Q_2$. For the individual tubes we have

\begin{eqnarray}
\Delta p&=&R_1\,Q_1\\
\Delta p&=&R_2\,Q_2
\end{eqnarray}

and for the combined parallel tube

\begin{eqnarray}
Q_1+Q_2&=&\frac{\Delta p}{R_1}+\frac{\Delta p}{R_2}\\
Q&=&\Delta p\left(\frac{1}{R_1}+\frac{1}{R_2}\right)=\left(\frac{R_1+R_2}{R_1\,R_2}\right)
\end{eqnarray}

and solving for $\Delta p$ we obtain:

$$\Delta p=\frac{R_1\,R_2}{R_1+R_2}\,Q\quad .$$

Thus the parallel resistance is 

$$R=\frac{R_1\,R_2}{R_1+R_2}\quad .\tag{15}$$


### Systems of Channels und Kirchhoff's law

Given the analog between the Hagen-Poiseuille law and Ohm's law, we can apply the Kirchhoffs's law for resistive networks also to networks of microchannels. An example of this approach is shown in the figure below.

<img src="pics/resistive_network.png" width=600px>

Channels with a hydraulic resistance $R_\text{hyd}$ become resistors, flow rates $Q$ become currents, and pumps delivering pressure differences $\Delta p$ become batteries.

1. The sum of flow rates entering/leaving any node in the circuit is zero.
2. The sum of all pressure differences in any closed loop of the circuit is zero.

### Viscous dissipation in time-dependent flows

We first derive an expression for the energy dissipation of a flow in a channels using the following Gedanken experiment (see figure below).

<img src="pics/viscous_dissipation.png" width=500px>

At time $t=0$ the driving force of the flow (the pressure difference $\Delta p$) is suddenly removed and the flow decellerates for $t\rightarrow\infty$ to rest. Let us now calculate the rate of change of the kinetic energy of the fluid assuming an incompressible flow and ignore gravity. The kinetic energy of the fluid occuyping the volume $\Omega$ is

$$E_{\text{kin}}=\int_\Omega \frac{1}{2}\rho\,u^2\,\mathrm{d}V=\int_\Omega \frac{1}{2}\rho\,\sum_{i=1}^3 u_i^2\,\mathrm{d}V=\int_\Omega \frac{1}{2}\rho\,u_i\,u_i\,\mathrm{d}V\tag{1}\quad$$
where for the last integral we used the Einstein summation rule. Now we look at the energy dissipation, i.e. the rate of change of the kinetic energy which is converted into heat (dissipated into heat). Then we obtain

$$\pad{E_{\mathrm{kin}}}{t}=\int_\Omega\rho\sum_{i=1}^3\,u_i\pad{u_i}{t}\,\mathrm{d}V=\int_\Omega\rho\,u_i\pad{u_i}{t}\,\mathrm{d}V\tag{2}$$

and again the Einstein summation rule is used in the right expression of Eq.(2). The Navier Stokes Equation provides the $\pad{\vb{u}_i}{t}
$ term as

$$
\rho\pad{u_i}{t}=-\rho\,u_j\pad{u_i}{x_j}+\mu\pad{}{x_j}\pad{}{x_j}u_i
\quad .\tag{3}$$

Why is the gradient of the pressure in the N.S. equation neglected? 

Inserting Eq.(3) into Eq.(2) results in 

$$\pad{E_{\mathrm{kin}}}{t}=\int_\Omega -\rho u_i u_j\pad{u_i}{x_j}+\mu u_i\pad{}{x_j}\pad{}{x_j}u_i\,\mathrm{d}V\tag{4}$$
Note that in Eq.(4) we use only the Einstein summation rule, i.e. sum over repeating indexes $i$ and $j$.

It can be shown (your exercise) that Eq. (4) is identical to

$$\pad{E_{\mathrm{kin}}}{t}=
\int_\Omega -\pad{}{x_j}\left( u_j\left[\frac{1}{2}\rho \,u_i\,u_i\right]-\mu \,u_i \pad{u_i}{x_j}\right)-\mu\left(\pad{u_i}{x_j}\right) \left(\pad{u_i}{x_j}\right)\,\mathrm{d}V\quad.\tag{5}$$


This seemingly more complex expressions allows us to apply the Gauss divergence theorem on the first term of the kernel. 

$$\pad{E_{\mathrm{kin}}}{t}=-
\int_{\partial\Omega} n_j\left( u_j\left[\frac{1}{2}\rho \,u_i\,u_i\right]-\mu \,u_i \pad{u_i}{x_j}\right)\,\mathrm{dA}-\mu \int_\Omega \left( \pad{u_i}{x_j} \right) \left( \pad{u_i}{x_j} \right) \,\mathrm{d}V\tag{6}$$

Equation 6 contains an integral over the surface in the first term $\partial \Omega$ and an integral over the volume $\Omega$ in the second term. Let us have a look at the surface first. The figure below depicts the 3 surfaces: the inlet $\partial \Omega_1$, the outlet $\partial \Omega_2$, and the solid wall $\partial \Omega_\text{wall}$.

<img src="pics/viscous_dissipation_sketch.png" width=450px>

At the solid wall $\partial \Omega_\text{wall}$ the velocity $u_i=0$, thus the integral for $\partial\Omega_\text{wall}$ is zero. The two remaining walls cancel each other. The reason is the translational invariance of the flow that makes the expression in the round brackets of the first integral idential for the two end surfaces. The two normal vectors point into two opposite directions thus canceling their contributions to the integral, see above figure.

Thus we have 

$$\pad{E_{\mathrm{kin}}}{t}=-\mu \int_\Omega \left( \pad{u_i}{x_j} \right) \left( \pad{u_i}{x_j} \right) \,\mathrm{d}V\tag{7}$$

If the flow is a planar Poiseuille flow, i.e. $\vb{u}=u_x(y,z)\,\vb{e}_x$, Eq. (7) can be simplified

$$\pad{E_{\mathrm{kin}}}{t}=-\mu \int_\Omega \left( \pad{u_x}{y} \right)^2 +\left( \pad{u_x}{z} \right)^2 \,\mathrm{d}V\tag{8}$$

Because $\mu$ is a positive constant and the integral is always positive the kinetic energy is decreasing. Let $W_\text{visc}$ be the heat generated (the work done), i.e. being a positive quantity, we can write

$$\pad{W_{\mathrm{visc}}}{t}=-\pad{E_{\mathrm{kin}}}{t}=\mu \int_\Omega \left( \pad{u_x}{y} \right)^2 +\left( \pad{u_x}{z} \right)^2 \,\mathrm{d}V\tag{8}$$

### Viscous dissipation for steady state flows


In the steady state $\pad{E_\text{kin}}{t}=0$, the mass flow through $\partial\Omega_1$ and $\partial\Omega_2$ is the same. This is achieved with a driving pressure difference $\Delta p$ between the two surfaces. The pressure relation is

$$p(\partial\Omega_1)=p(\partial\Omega_1)+\Delta p\quad .$$

Because the kinetic energy of the fluid is constant he rate of viscous heat dissipation is now balance by the mechanical power from the driving pressure force 

$$\pad{E_{\mathrm{kin}}}{t}=\pad{W_\text{mech}}{t}-\pad{W_{\mathrm{visc}}}{t}=0\tag{9}\quad .$$

We can now repeat with Eq. (2), knowing this is 0 and insert the Navier Stokes equation (3), which however has now the pressure term.

$$
\rho\pad{u_i}{t}=-\rho\,u_j\pad{u_i}{x_j}+\mu\pad{}{x_j}\pad{}{x_j}u_i-\pad{p}{x_i}
\quad .\tag{10}$$

The first two terms of Eq. (10) have been used to determine $-\pad{W_\text{visc}}{t}$ and thus 

$$\pad{W_\text{mech}}{t}=\int_\Omega u_i\left(-\pad{p}{x_i}
\right)\,\mathrm{d}V=-\int_\Omega \pad{(u_i \,p)}{x_i}\,\mathrm{d}V=-\int_{\partial\Omega} n_i\left(u_i \,p\right)\,\mathrm{d}A\quad .\tag{11}
$$

We haves used in Eq. (11) again the Gauss Divergence theorem to convert the volume integral into a surface integral. This last integral can be evaluated because the pressure on the end face is constant  and the velocity distribution is the same on both faces, i.e. translational invariance:

$$\pad{W_\text{mech}}{t}=p(\partial\Omega_1)\int_{\partial\Omega_1} u_x\,\mathrm{d}A-p(\partial\Omega_2)\int_{\partial\Omega_2} u_x\,\mathrm{d}A=\Delta p\int_{\partial\Omega_1}u_x(y,z)\,\mathrm{d}A=Q\Delta p\tag{12}$$

The signs are a result from the direction of the normal vector $\vb{n}$, namely $\vb{n}(\partial \Omega_1)=-\vb{e}_x=-\vb{n}(\partial \Omega_2)$.

Thus we can connect now the pressure loss in microchannel with the flow field

$$\pad{W_\text{mech}}{t}=\pad{W_\text{vis}}{t}=\mu \int_\Omega \left( \pad{u_x}{y} \right)^2 +\left( \pad{u_x}{z} \right)^2 \,\mathrm{d}V=Q\Delta p\tag{13}\quad .$$
