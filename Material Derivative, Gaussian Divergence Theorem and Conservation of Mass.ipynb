# Material Derivative


The material derivative is the derivative with respect to time of a physical quantity of a fluid particle described by field variables which are fixed in space.

This sounds rather complicated, but isn't. Consider you want to measure the velocity field of a fluid. Most conveniently you fix the coordinate system with the lab and measure at fixed points in this coordinate system the velocity field. You thus do not measure the velocity of a single fluid particle moving through the fluid but the velocity at a specific point.

How do we relate this measurement for example to the acceleration a fluid particle undergoes? If the flow is unsteady, i.e. the fluid velocity is time dependent, then you may want to define an acceleration of the fluid particle at postion $\vb{x}$ with 

$$\vb{a}=\pad{\vb{u}(\vb{x},t)}{t}\tag{1}$$.

Is that all of the acceleration a fluid particle will experience?
Before you answer this question let's have a look at the following steady (time independent) flow.

![image](pics/md_fig1.png)
<p>
Figure 1: Acceleration of fluid in a nozzle.


This is a called a fluid constriction. The fluid enters with a constant velocity from the left and exits on the right. Assume the fluid is incompressible, then the volume of fluid entering on the left must be equal to the volume on the right. Thus the volumetric flow rate is fixed over the whole configuration. The fluid entering from the left has a smaller velocity than exciting on the right. Therefore, after a fluid particle enters it will be accelerated while moving through the constriction. Remember, the fluid velocity *is not* a function of time, thus our acceleration as (wrongly) defined above would give us 0. 

To do it right, we need to calculate the acceleration of a fluid particle differently, and not only account for the change of the velocity in time but also in space. For above fluid constriction, the geometry leads to an acceleration.

Now we derive in a formal way the time derivative of a fluid particle by using only quantities which can be obtained from a fixed lab based coordinate system. Describing a flow with a lab based coordinate systems is termed *Eulerian description*. In contrast if we fix the fluid particle and describe its change in time we use the technical term *Lagrangian description*.

Before we look at the change of the velocity with space and time, we make our life easier by looking at a scalar quantity, e.g. the density $\rho(\vb{x},t)$. Consider we have measured the density of the fluid at every point in the flow using a lab based coordinate system.
We now want to know the change of the density of a fluid particle while it moves. 

Let us first identify  this fluid particle... that is easy, it is the fluid particle being at time $t=t_0$ at position $\vb{x}=\vb{x_0}=(x_{10},x_{20},x_{30})^T$. 

<div class="YourWork">**Why is this unique?**</div>

The notation of the change of the fluid particles density is:

$$\left.\fudm{\rho(\vb{x},t)}{t}\right|_{\vb{x_0},t_0}\quad .\tag{2}$$

This is called the *material derivative*. The capital D is used to emphasize that this derivative is taken for a moving fluid particle. A formal way to define the material derivative is using the limit of the difference quotient, i.e. 

$$\left.\fudm{\rho(\vb{x},t)}{t}\right|_{\vb{x_0},t_0}=
\lim_{(t-t_0)\rightarrow 0}
\frac{\rho(\vb{x},t)-\rho(\vb{x_0},t_0)}{t-t_0}\tag{3}$$

As we know $\rho(\vb{x_0},t_0)$, we would only need to obtain $\rho(\vb{x},t)$ to calculate this limit. This can be obtained with a Taylor expansion of the flow field around $\vb{x_0}$. Let us conduct this Taylor approximation to first order and insert it into Eq. (3). 


\begin{eqnarray*}
\rho(\vb{x},t)\approx \rho(\vb{x_0},t_0) &+& (t-t_0) \left.\left(\pad{\rho}{t} \right)\right|_{\vb{x_0},t_0}+(x_1-x_{10}) \left.\left(\pad{\rho}{x_1} \right)\right|_{\vb{x_0},t_0}\\
&+& (x_2-x_{20}) \left.\left(\pad{\rho}{x_2} \right)\right|_{\vb{x_0},t_0} +
(x_3-x_{30}) \left.\left(\pad{\rho}{x_3} \right)\right|_{\vb{x_0},t_0}
\end{eqnarray*}

By moving $\rho(\vb{x_0},t_0)$ to the L.H.S and dividing both sides by $t-t_0$ we obtain:

\begin{eqnarray*}
\frac{\rho(\vb{x},t)-\rho(\vb{x_0},t_0)}{t-t_0} &\approx& \left.\left(\pad{\rho}{t} \right)\right|_{\vb{x_0},t_0}+\frac{x_1-x_{10}}{t-t_0}\left.\left(\pad{\rho}{x_1} \right)\right|_{\vb{x_0},t_0}\\
&+& \frac{x_2-x_{20}}{t-t_0} \left.\left(\pad{\rho}{x_2} \right)\right|_{\vb{x_0},t_0} +
\frac{x_3-x_{30}}{t-t_0}\left.\left(\pad{\rho}{x_3} \right)\right|_{\vb{x_0},t_0}
\end{eqnarray*}

Now we need to take the limit $(t-t_0)\rightarrow 0$. The ratios on the R.H.S. under this limit become identical to the velocities $u_1$, $u_2$, and $u_3$, respectively.
Thus we can write

\begin{eqnarray*}
\fudm{\rho(\vb{x},t)}{t}&=&\pad{\rho}{t}+u_1\pad{\rho}{x_1}+u_2\pad{\rho}{x_2}+u_3\pad{\rho}{x_3}\\
\fudm{\rho(\vb{x},t)}{t}&=&\pad{\rho}{t}+\vb{u}\cdot \nabla \rho
\end{eqnarray*}

Thus the temporal change of the density of a fluid (read Lagrangian) particle $\fudm{\rho}{t}$ results from two contributions, the temporal change of density at an Eularian point $\pad{\rho}{t}$ and the spatial variation $\vb{u} \cdot \nabla \rho$.

We have now derived the material derivative of scalar quantity, e.g. $\rho$, but we started on top with our example with the acceleration of the fluid particles, which is a vector quantity.
If we would naively insert $\vb{a}$ in the definition of the material derivative we would have a to deal with the term $\vb{u} \cdot \nabla \vb{u}$, a gradient of a vector. What is this?

Let's go step by step, we use the definition for the individual components of the velocity vector we obtain these three equations:

\begin{eqnarray*}
\fudm{u_1(\vb{x},t)}{t}&=&\pad{u_1}{t}+\vb{u}\cdot \nabla u_1\\
\fudm{u_2(\vb{x},t)}{t}&=&\pad{u_2}{t}+\vb{u}\cdot \nabla u_2\\
\fudm{u_3(\vb{x},t)}{t}&=&\pad{u_3}{t}+\vb{u}\cdot \nabla u_3
\end{eqnarray*}

Before we write above in a single equation we'll have a look at the first equation in component form:

$$
\fudm{u_1(\vb{x},t)}{t}=\pad{u_1}{t}+ u_1\pad{u_1}{x_1}+u_2\pad{u_1}{x_2}+u_3\pad{u_1}{x_3}\tag{4}
$$

Or using the Einstein notation:

$$\fudm{u_i}{t}=\pad{u_i}{t}+u_j\pad{u_i}{x_j}\tag{5}
$$

The second term on the R.H.S is abbreviated as $\vb{u}\cdot\nabla\vb{u}$ or $(\vb{u}\cdot\nabla)\vb{u}$ and is only valid in this form for a cartesian coordinate system. Some call this term "u grad u" or "u del u". Both is fine, please remember that is stands for this particular operator. 
The acceleration of a fluid particle in short form is:

$$
\fudm{\vb{u}(\vb{x},t)}{t}=\pad{\vb{u}}{t}+(\vb{u}\cdot \nabla) \vb{u}\tag{6}
$$

# Conservation of Mass

## Gaussian divergence theorem


Consider a vector field $\vb{h}(\vb{x})$

$$\vb{h}=\begin{pmatrix}h_1(x_1,x_2)\\h_2(x_1,x_2)\end{pmatrix}\tag{1}$$

and the component in normal direction is given by 

$$h_n=\vb{h}\cdot\vb{n}=h_1 n_1 + h_2 n_2\tag{2}$$

with the definition of the divergence

$$\nabla \cdot \vb{h}= \pad{h_1}{x_1} + \pad{h_2}{x_2}\tag{3}$$

then the *Gaussian divergence theorem* states that the integral over the closed loop along contour $\cal C$ is 

$$\oint_{\cal C} \vb{h}\cdot\vb{n} \,\dd l=
\int \int_{\cal D} \nabla \cdot \vb{h} \, \dd x_1 \, \dd x_2\tag{4}\quad . $$

The Gaussian divergence theorem is very helpful in formulating conservation laws. This becomes immediately visible if we use the flow field $\vb{u}$ as the vector field $\vb{h}$.


## Conservation of Mass

For the conservation of mass we need to define the mass flux $q_m=\rho u_n$ and the integral over a line/area is the mass flow rate

$$Q_m=\int_A^B q_m \dd l = \int_A^B \rho \vb{u}\cdot\vb{n} dl=
\int_A^B \rho \left(u_1 \dd x_2 - u_2 \dd x_1\right)\quad .\tag{6}$$ 

To apply the divergence theorem we need to form a closed surface area as shown in Fig. 1.

\begin{eqnarray}
Q_m&=&\oint q_m \dd l =\oint \rho \left(u_1 \dd x_2 - u_2 \dd x_1\right)\tag{8a}\\
&=& \int\int_{\cal D} \nabla \cdot \left(\vb{\rho}\vb{u}\right) \dd x_1 \dd x_2\tag{8b}\\
&=& \int\int_{\cal D} \left( \vb{u}\cdot\nabla \rho + \rho\nabla \cdot \vb{u}\right)\dd x_1 \dd x_2\tag{8c}
\end{eqnarray}

In Eq. (8b) we have applied the Gaussian divergence theorem and in Eq. (8c) the chain rule.

If the fluid is compressible, than not necessarily the mass of the domain $\cal D$ remains constant. The change of mass in domain $\cal D$ is equal to the mass flux. The mass is calculated by the area/volume integral of the density. Let us write this idea as a formula:

$$\fud{}{t}\int \int \rho \dd x_1 \dd x_2=-Q_m=-\int \int_{\cal D} \nabla \cdot (\rho \vb{u}) \,\dd x_1 \dd x_2\tag{9}$$

The domain $\cal D$ is fixed in space, therefore the time derivative does not affect the boundaries and Equation (9) can be written as a single kernel:

$$\int \int \pad{\rho}{t} +\nabla \cdot (\rho \vb{u})\,\dd x_1 \dd x_2=0\tag{10}\quad .$$

Now an interesting argument is used. The domain $\cal D$ has not specific properties besides being enclosed by a loop $\cal C$. As the domain has not particular properties and can be applied to any part of the flow, the kernel itself must be zero. Thus we can demand that 

$$\pad{\rho}{t} +\nabla \cdot (\rho \vb{u})=0\tag{11a}$$

or

$$\pad{\rho}{t} +\nabla \rho \cdot \vb{u} + \rho \, \nabla \cdot \vb{u}=0\tag{11b}\quad .$$

With the [definition](03_Material%20Derivative.ipynb) of the *material derivative* 
we can rewrite Eq. (11a) as 

$$\fudm{\rho}{t}+\rho\,\nabla\cdot\vb{u}=0\tag{12}\quad .$$

Looking at Eq. (12) the definition of incompressible flow is straight forward. An incompressible fluid particle is a fluid particle who's density does not change, i.e. $\fudm{\rho}{t}=0$. If this is zero, $\nabla \cdot \vb{u}=0$ must hold.
